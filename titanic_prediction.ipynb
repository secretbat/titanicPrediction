```python
X = df1.iloc[:,1:]
y = df1.iloc[:,0]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=20)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
```

```python
param_distributions = {
    'n_estimators' : [int(x) for x in np.linspace(start=100,stop=1200,num=12)],
    'criterion' : ['gini','entropy'],
    'max_depth' : [int(x) for x in np.linspace(3,30,num=10)],
    'min_samples_split' : [2,5,10,15,20,25,30],
    'min_samples_leaf' : [1,2,5,10,15]
}
print(param_distributions)
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
rf_random = RandomizedSearchCV(estimator=rf,param_distributions=param_distributions,n_iter=10,cv=5,random_state=20)
```

```python
rf_random.fit(X_train,y_train)
```

```python
print(rf_random.best_params_)
```

```python
rf = RandomForestClassifier(n_estimators=1200,criterion='entropy',max_depth=3,min_samples_split=30,min_samples_leaf=2)
rf.fit(X_train,y_train)
predictions = rf.predict(X_test)
```

```python
print(list(y_test))
print(predictions)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
accuracy_score(y_test,predictions)
confusion_matrix(y_test,predictions)
```
